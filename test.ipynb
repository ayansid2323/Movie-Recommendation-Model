```python
import pandas as pd
movies = pd.read_csv('./data/movies.csv', sep=',')
movies = movies[:100]
```

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer

def custom_tokenizer(text):
    """
    Tokenizes the input text into a list of words, excluding any words that contain digits.

    Args:
        text (str): The input text to be tokenized.

    Returns:
        list: A list of tokens (words) from the input text, excluding any words that contain digits.
    """
    tokens = re.findall(r'\b\w+\b', text)
    return [token for token in tokens if not any(char.isdigit() for char in token)]

vectorizer = TfidfVectorizer(stop_words='english', tokenizer=custom_tokenizer)
X_train = vectorizer.fit_transform(movies["overview"])
feature_names = vectorizer.get_feature_names_out()
```

```python
pd.DataFrame(X_train.toarray(), columns=feature_names).head(6)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix0 = cosine_similarity(X_train, X_train)
similarity_matrix0[:5,:5]
```

```python
from sklearn.metrics.pairwise import linear_kernel
similarity_matrix1 = linear_kernel(X_train, X_train)
similarity_matrix1[:5,:5]
```

```python
movie_name = "Spectre"
movie_idx = movies[movies["title"]==movie_name].index[0]
print(f"Movie index for '{movie_name}': {movie_idx}")
```

```python
similarity_scores = list(enumerate(similarity_matrix0[movie_idx]))
similarity_scores_sorted = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
similarity_scores_sorted[:5][1:]
```

```python
recommended_movie_indices = [i for i, _ in similarity_scores_sorted[:5][1:]]
recommended_movie_indices
recommended_movies = list(movies["title"].iloc[recommended_movie_indices])
for movie in recommended_movies:
    print(movie)
```
